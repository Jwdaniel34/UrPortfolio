```python
import bs4 as BeautifulSoup
import pandas as pd
import re
import requests
import time
import os
import os.path
from selenium import webdriver
from tqdm import tqdm
url = 'http://www.stockpup.com/data/'
```

```python
def get_stock_ticker(stringg):
    final = ''
    stringg = stringg.strip('<a href="/data/')
    for char in stringg[:4]:
        if char.isalpha() == True and char == char.title():
            final += char
    return final
```

```python
ins = webdriver.Chrome()
ins.get(url)
source = BeautifulSoup.BeautifulSoup(ins.page_source)
div = source.find_all('div', {'class':'col-md-4 col-md-offset-1'})
all_as = div[0].find_all('a')
```

```python
href = ''
for i in tqdm(range(len(all_as))):
    if 'CSV' in all_as[i].text:
        href = all_as[i]['href']
        dirty_string = str(all_as[i])
        ticker = get_stock_ticker(dirty_string)
        source = f'/Users/flatironschool/Downloads/{href[6:]}'
        destination = f'/Users/flatironschool/DS071519/fundamental_stock_classifier/stock_csvs/stock_pup_{ticker}.csv'
        if os.path.isfile(destination):
            continue
        ins.get('http://www.stockpup.com/{}'.format(href))
        time.sleep(4)
        while(1):
            try:
                os.rename(source, destination)
                break
            except FileNotFoundError:
                time.sleep(1)
                continue
    else:
        pass
        
```

```python
ins.quit()
```

```python
tickers = []
for i in range(0,len(all_as),2):
    dirty_string = str(all_as[i])
    ticker = get_stock_ticker(dirty_string)
    tickers.append(ticker)
```

```python
pd.DataFrame(tickers).to_csv('/Users/flatironschool/DS071519/fundamental_stock_classifier/tickers.csv', index = False)
```
