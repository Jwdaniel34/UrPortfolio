```python
tickers = pd.read_csv('tickers.csv')
ticker_list = tickers['0'].values.tolist()
```

```python
new_companies = []
no_finds = []
for ticker in ticker_list:
    try:
        url = f'https://finviz.com/quote.ashx?t={ticker}&ty=c&p=d&b=1'
        page = requests.get(url)
        print(page)
        soup = BeautifulSoup.BeautifulSoup(page.content, 'html.parser')
        company = soup.select("tr  a[class=tab-link] b")[1].text
        sector = soup.select("td  a[class=tab-link] ")[10].text
        beta = soup.select('tr td[align=left] b')[44].text
        dict1 = {'ticker': ticker, 'company': company, 'sector': sector, 'beta': beta}
        new_companies.append(dict1)
    except:
        print(f'ticker {ticker}')
        no_finds.append(ticker)
```

```python
new_companies
```

```python
tickers = tickers.rename(columns = {'Ticker': 'ticker'})
```

```python
ticks = pd.DataFrame.from_dict(new_companies)
```

```python
ticker_new = pd.merge(tickers, ticks, on='ticker', how='outer', suffixes=('_made', '_close'))
```

```python
ticker_new['sector'] = ticker_new['sector'].fillna(value='Misc')
```

```python
ticker_new['company'] = ticker_new['company'].fillna(value='nomatch')
```
